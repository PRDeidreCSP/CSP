```python

#practicas de tipo de datos y entradas de usuario
#crear un programa que calcule el indice de masa corporal de una persona



print("vamos al calcular tu masa corporal")

peso= input("cual es tu peso en kg?: ")
estatura= input("cual es tu estatura en metros? ")
imc= float(peso) / (float(estatura) ** 2)                
print("tu indice de masa corporal es: " + str(imc))
```
